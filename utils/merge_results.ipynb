```python
import pandas as pd
import os

def round_floats_in_dicts(list_of_dicts):
    rounded_list = []
    for original_dict in list_of_dicts:
        new_dict = {}
        for key, value in original_dict.items():
            if isinstance(value, float):
                new_dict[key] = round(value, 3)
            else:
                new_dict[key] = value
        rounded_list.append(new_dict)
    return rounded_list
```
